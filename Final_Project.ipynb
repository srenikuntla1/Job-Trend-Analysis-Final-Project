```python
# create a machine learning model to predict how many student in the future will do information technology related jobs
# import the necessary packages

from sklearn import preprocessing
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# create a label encoder object
le = preprocessing.LabelEncoder()

# convert the categorical columns into numeric
df['current_job_category'] = le.fit_transform(df['current_job_category'])
df['current_job_title'] = le.fit_transform(df['current_job_title'])
df['current_employer_industry'] = le.fit_transform(df['current_employer_industry'])

# create a dataframe with only the columns that are required for the machine learning model
df1 = df[['current_job_category', 'current_job_title', 'current_employer_industry']]
df1
```

```python
# machine learning model to predict the job category of a student in the future from df1
# create a feature set
X = df1.values[:, 1:3]
X

# create a target set
Y = df1.values[:, 0]
Y

# split the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

# create a decision tree classifier object
dt = tree.DecisionTreeClassifier()

# train the model using the training sets
dt.fit(X_train, y_train)

# predict the response for test dataset
y_pred = dt.predict(X_test)
y_pred

# model accuracy, how often is the classifier correct?
print("Accuracy:", accuracy_score(y_test, y_pred))

# confusion matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))

# classification report
print("Classification Report:\n", classification_report(y_test, y_pred))
```
