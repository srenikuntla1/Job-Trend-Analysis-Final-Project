```python
#convert csv file to data frame using pandas
import pandas as pd
df = pd.read_csv('SteppingBlocksData.csv')
df
```

```python
df.shape
```

```python
# remove rows having current_job_title as 'NaN' and current_job_category as 'NaN'
df = df.dropna(subset=['current_job_title', 'current_job_category'], how='all')
df
```

```python
df.shape
```

```python
# print all the unique values in the column 'current_job_category'
df['current_job_category'].unique()
# print them as a dataframe
df['current_job_category'].unique()
# create a dataframe with the unique values in the column 'current_job_category'
current_job_category = pd.DataFrame(df['current_job_category'].unique())
# name the columen 'current_job_category'
current_job_category.columns = ['current_job_category']
current_job_category
```

```python
# print all the unique values in the column 'current_job_title'
df['current_job_title'].unique()
# print them as a dataframe
df['current_job_title'].unique().tolist()
#create a daraframe named Job_title and store the unique values of current_job_title in it
Job_title = df['current_job_title'].unique()
Job_title = pd.DataFrame(Job_title)
# name the column 'current_job_title'
Job_title.columns = ['current_job_title']
Job_title
```

```python
# create a dictionary with key as a 'current_job_category' and value as a list of 'current_job_title'
category_title = {}
for i in current_job_category['current_job_category']:
    category_title[i] = df[df['current_job_category'] == i]['current_job_title'].unique().tolist()
category_title
```
